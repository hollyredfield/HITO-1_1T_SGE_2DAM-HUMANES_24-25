```python
import pandas as pd
import requests
from IPython.display import display
```

```python
# Descargar el archivo Excel del INE
url = "https://www.ine.es/jaxiT3/files/t/es/xlsx/62271.xlsx?nocab=1"
file_path = "dataset.xlsx"

# Descargar el archivo y guardarlo localmente
response = requests.get(url)
with open(file_path, 'wb') as file:
    file.write(response.content)

# Leer el archivo Excel en un DataFrame
df = pd.read_excel(file_path, engine='openpyxl')
print("Datos cargados en el dataset")
```

```python

# Función para limpiar datos
def limpiar_datos(df):
    # Reemplazar valores NaN por una cadena vacía
    df.fillna('', inplace=True)
    # Eliminar columnas sin nombre
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    return df

df = limpiar_datos(df)
print("Datos limpiados")

# Mostrar las primeras filas del dataset limpio
display(df)
```

```python
# Guardar los datos en un archivo de texto
with open('lista.txt', 'w') as f:
    for row in df.values:
        f.write(' '.join(map(str, row)) + '\n')
print("Datos guardados en lista.txt")
```

```python
# Función para calcular estadísticas
def calcular_estadisticas(df):
    # Filtrar solo las columnas numéricas
    df_numerico = df.select_dtypes(include='number')
    moda = df_numerico.mode()
    estadisticas = {
        'media': df_numerico.mean(),
        'varianza': df_numerico.var(),
        'moda': moda.iloc[0] if not moda.empty else 'No hay moda'
    }
    return estadisticas

estadisticas = calcular_estadisticas(df)
print("Estadísticas calculadas")
print(estadisticas)
```

```python
# Clase Dataset
class Dataset:
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
    
    def __str__(self):
        return str(self.__dict__)
    
    def actualizar_atributo(self, atributo, valor):
        setattr(self, atributo, valor)
    
    def __eq__(self, otro):
        return self.columna2 == otro.columna2
    
    def __add__(self, otro):
        return Dataset(**{k: v + otro.__dict__[k] for k, v in self.__dict__.items()})
    
    def __sub__(self, otro):
        return Dataset(**{k: v - otro.__dict__[k] for k, v in self.__dict__.items()})

# Crear objetos y probar métodos
obj1 = Dataset(columna1=1, columna2=2)
obj2 = Dataset(columna1=3, columna2=4)
print(obj1 + obj2)
print(obj1 - obj2)

# Mostrar la tabla final
display(df)
```
